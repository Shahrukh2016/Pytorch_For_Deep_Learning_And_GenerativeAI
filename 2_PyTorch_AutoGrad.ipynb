### 1. Example:
- y = x ** 2

```python
x = torch.tensor(data = 3.0, requires_grad=True)
y = x ** 2

print(x)
print(y)
```

### 2. Example:
- x = y ** 2
- z = Sin(y)

```python
x = torch.tensor(data = 3.0, requires_grad=True)
y = x ** 2
z = torch.sin(y)

print(x)
print(y)
print(z)
```

```python
## calculating derivatives
z.backward()

## checking derivative of z wrt x
print(x.grad)
```

### 3. Example:
- z = w*x + b
- sig = 1/(1-e*-z)

```python
# Binary Cross-Entropy Loss for scalar
def binary_cross_entropy_loss(prediction, target):
    epsilon = 1e-8  # To prevent log(0)
    prediction = torch.clamp(prediction, epsilon, 1 - epsilon)
    return -(target * torch.log(prediction) + (1 - target) * torch.log(1 - prediction))
```

```python
x = torch.tensor(6.7)
y = torch.tensor(0.0)

print(x,y)
```

```python
w = torch.tensor(1.0, requires_grad=True)
b = torch.tensor(0.0, requires_grad=True)

print(w, b)
```

```python
## defining functions
z = w*x + b
print(z)

y_pred = torch.sigmoid(z)
print(y_pred)
```

```python
loss = binary_cross_entropy_loss(y_pred, y)
loss
```

```python
loss.backward()
```

```python
print(w.grad)
print(b.grad)
```

### Working with Vector Inputs

```python
## defining tensor
x = torch.tensor(data = [1.0, 2.0, 3.0], requires_grad=True)
print(x)
```

```python
## Mapping / Function
y = (x**2).mean()
print(y)
```

```python
## calculating gradiants
y.backward()
```

```python
## checking gradiants
x.grad
```

### Clearning gradiants

```python
## defining tensor
x = torch.tensor(data = 3, requires_grad=True, dtype = torch.float32)
print(x)
```

```python
## Mapping / Function
y = (x**2).mean()
print(y)
```

```python
## calculating gradiants
y.backward()
```

```python
## checking gradiants
x.grad
```

```python
## since the gradient is changing on each forward and backward pass, we need to clear the gradients
x.grad.zero_()
```

### Disabling AutoGrad

```python
## while traing the NN we need to enable the AutoGrad as we need to calculate the gradients, but once we are done with training and we need only predictions, we should disable the AutoGrad to maintain the memory.

# option1: requires_grad_(False)
# option2: detach()
# option3: torch.no_grad()
```

#### Method 1: requires_grad_(False)


```python
## defining tensor
x = torch.tensor(data = 3, requires_grad=True, dtype = torch.float32)
print(x)
```

```python
## disabling the grad
x.requires_grad_(False)
```

```python
## Mapping / Function
y = (x**2).mean()
print(y)
```

```python
## calculating gradiants
y.backward()
```

#### Method 2: detach()


```python
## defining tensor
x = torch.tensor(data = 3, requires_grad=True, dtype = torch.float32)
print(x)
```

```python
## disabling the grad
z = x.detach()
print(z)
```

```python
## Mapping / Function
y = (z**2).mean()
print(y)
```

```python
## calculating gradiants
y.backward()
```

#### Method 3: torch.no_grad()

```python
## defining tensor
x = torch.tensor(data = 3, requires_grad=True, dtype = torch.float32)
print(x)
```

```python
## Mapping / Function
with torch.no_grad():
  y = (x**2).mean()
```

```python
y.backward()
```
